```python
#yolo_detect_eye_saves_crop
import os
import cv2
from ultralytics import YOLO

# Paths
IMAGE_DIR = "/home/khushi/Pixonate/Vit_conjuctiva/Pred_jan15onwards/p/Conjuctiva"
OUTPUT_DIR = "/home/khushi/Pixonate/Vit_conjuctiva/Pred_jan15onwards/p/test_eye_crops"
MODEL_PATH = "/home/khushi/Pixonate/New_Anemia/Models/Finetuned model/Eye model/Fbest.pt"   # your trained YOLO model

os.makedirs(OUTPUT_DIR, exist_ok=True)

# Load YOLO model
model = YOLO(MODEL_PATH)

for img_name in os.listdir(IMAGE_DIR):
    if not img_name.lower().endswith((".jpg", ".png", ".jpeg")):
        continue

    img_path = os.path.join(IMAGE_DIR, img_name)
    image = cv2.imread(img_path)

    if image is None:
        continue

    # Run YOLO
    results = model(image)[0]

    if len(results.boxes) == 0:
        continue  # no eye detected

    # Take highest confidence detection
    boxes = results.boxes
    best_box = boxes[boxes.conf.argmax()]

    x1, y1, x2, y2 = map(int, best_box.xyxy[0])

    # Crop eye
    eye_crop = image[y1:y2, x1:x2]

    if eye_crop.size == 0:
        continue

    # Save crop
    save_path = os.path.join(OUTPUT_DIR, img_name)
    cv2.imwrite(save_path, eye_crop)

print("✅ Eye cropping completed.")
```

```python
#creates_csv_of_hb_&_img-names
import os
import pandas as pd

IMAGE_DIR = "eye_crops"
OUTPUT_CSV = "image_hb.csv"

data = []

for img_name in os.listdir(IMAGE_DIR):
    if not img_name.lower().endswith((".jpg", ".png", ".jpeg")):
        continue

    name_without_ext = os.path.splitext(img_name)[0]  # e.g. 3.11.5
    parts = name_without_ext.split(".")

    if len(parts) < 3:
        continue  # safety check

    # Extract Hb value
    hb_value = float(parts[1] + "." + parts[2])

    data.append({
        "image_name": img_name,
        "hb_value": hb_value
    })

df = pd.DataFrame(data)
df.to_csv(OUTPUT_CSV, index=False)

print("✅ CSV created:", OUTPUT_CSV)
```
