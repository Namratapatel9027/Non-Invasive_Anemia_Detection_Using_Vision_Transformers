```python
#copied_matched_images_based_on_original_csv
import pandas as pd
import os
import shutil

# Paths
csv_path = r"/home/khushi/Pixonate/New_Anemia/R_new_anemia/AnemiaData_from_2025_11_24.csv"
upload_folder = r"/home/khushi/Pixonate/New_Anemia/R_new_anemia/Manually_downld"
output_base_folder = r"/home/khushi/Pixonate/New_Anemia/R_new_anemia/copied_images"
output_csv = r"/home/khushi/Pixonate/New_Anemia/R_new_anemia/AnemiaData_with_copied_images.csv"

os.makedirs(output_base_folder, exist_ok=True)

# Read CSV
df = pd.read_csv(csv_path)
df['CopiedImagePaths'] = ""

# List all files in folder, lowercase for case-insensitive match
folder_files = {f.lower(): f for f in os.listdir(upload_folder)}

# Counters
total_images = 0
total_found = 0
total_copied = 0
total_skipped = 0

for row in df.itertuples():
    total_images += 1
    image_paths = str(row.ImagePath).split(",")
    image_types = str(row.Imagetype).split(",")

    copied_paths = []

    if len(image_paths) > 0 and len(image_types) > 0:
        first_img_csv = os.path.basename(image_paths[0].strip()).lower()  # lowercase
        first_type = image_types[0].strip().replace(" ", "_").replace(".", "_")

        # Check if filename exists in folder (case-insensitive)
        if first_img_csv in folder_files:
            src_file_actual = folder_files[first_img_csv]
            src_path = os.path.join(upload_folder, src_file_actual)
            type_folder = os.path.join(output_base_folder, first_type)
            os.makedirs(type_folder, exist_ok=True)
            dst_path = os.path.join(type_folder, src_file_actual)
            shutil.copy(src_path, dst_path)
            copied_paths.append(dst_path)
            total_found += 1
            total_copied += 1
        else:
            print(f"Skipped (not found in folder): {first_img_csv}")
            total_skipped += 1

    df.at[row.Index, 'CopiedImagePaths'] = ",".join(copied_paths)

# Save CSV
df.to_csv(output_csv, index=False)

# Summary
print("\nCopy Summary:")
print(f"Total first-position images in CSV: {total_images}")
print(f"Images found: {total_found}")
print(f"Images copied: {total_copied}")
print(f"Images skipped/missing: {total_skipped}")
```

```python
# copied_matched_images_based_on_original_csv_1
import pandas as pd
import os
import shutil

# Paths
csv_path = r"/home/khushi/Pixonate/New_Anemia/R_new_anemia/AnemiaData.csv"
upload_folder = r"/home/khushi/Pixonate/New_Anemia/R_new_anemia/overall_Main_dataset"
output_base_folder = r"/home/khushi/Pixonate/New_Anemia/R_new_anemia/overall_main_dataset_found"
output_csv = r"/home/khushi/Pixonate/New_Anemia/R_new_anemia/overall_data_with_copied_images.csv"

os.makedirs(output_base_folder, exist_ok=True)

# Read CSV
df = pd.read_csv(csv_path)

# Output columns
df["CopiedImagePaths"] = ""
df["NotCopiedImage"] = ""

# Case-insensitive lookup (DOES NOT change original names)
folder_files = {f.lower(): f for f in os.listdir(upload_folder)}

# Counters
total_rows = 0
total_images = 0
total_copied = 0
total_skipped = 0

for row in df.itertuples():
    total_rows += 1

    image_paths = str(row.ImagePath).split(",")
    image_types = str(row.Imagetype).split(",")

    copied_paths = []
    not_copied = []

    # Use first image type only (same logic as before)
    if len(image_types) > 0:
        type_folder = image_types[0].strip().replace(" ", "_").replace(".", "_")
        type_folder_path = os.path.join(output_base_folder, type_folder)
        os.makedirs(type_folder_path, exist_ok=True)
    else:
        continue

    # 🔁 LOOP THROUGH ALL IMAGES
    for img in image_paths:
        img_name = os.path.basename(img.strip())
        img_name_lower = img_name.lower()
        total_images += 1

        if img_name_lower in folder_files:
            actual_name = folder_files[img_name_lower]
            src_path = os.path.join(upload_folder, actual_name)
            dst_path = os.path.join(type_folder_path, actual_name)

            shutil.copy(src_path, dst_path)

            copied_paths.append(dst_path)
            total_copied += 1
        else:
            print(f"Skipped (not found): {img_name}")
            not_copied.append(img_name)
            total_skipped += 1

    # Write results to CSV
    df.at[row.Index, "CopiedImagePaths"] = ",".join(copied_paths)
    df.at[row.Index, "NotCopiedImage"] = ",".join(not_copied)

# Save CSV
df.to_csv(output_csv, index=False)

# Summary
print("\nCopy Summary:")
print(f"Total CSV rows processed: {total_rows}")
print(f"Total images checked: {total_images}")
print(f"Images copied: {total_copied}")
print(f"Images skipped/missing: {total_skipped}")
```

```python
import pandas as pd
import os

# Paths
csv1 = "/home/khushi/Pixonate/New_Anemia/R_new_anemia/AnemiaData_no_images_found.csv"
csv2 = "/home/khushi/Pixonate/New_Anemia/R_new_anemia/AnemiaData_with_copied_images_md_1.csv"

# Read CSVs
df_not_found = pd.read_csv(csv1)
df_copied = pd.read_csv(csv2)

# Collect all NOT copied image names (CSV 1)
not_found_images = set()

for val in df_not_found["NotCopiedImage"].dropna():
    for img in str(val).split(","):
        not_found_images.add(os.path.basename(img.strip()))

# Collect all COPIED image names (CSV 2)
copied_images = set()

for val in df_copied["CopiedImagePaths"].dropna():
    for img in str(val).split(","):
        copied_images.add(os.path.basename(img.strip()))

# Find intersection
found_again = not_found_images.intersection(copied_images)

# Result
print("Total not-found images that are found in copied images of 2nd CSV:", len(found_again))
```

```python
import pandas as pd
import os

# Paths
input_csv = "/home/khushi/Pixonate/New_Anemia/R_new_anemia/AnemiaData_no_images_found.csv"
check_folder = "/home/khushi/Pixonate/New_Anemia/R_new_anemia/1"
output_csv = "/home/khushi/Pixonate/New_Anemia/R_new_anemia/AnemiaData_recovered_images.csv"

# Read CSV
df = pd.read_csv(input_csv)

# New column
df["FoundInFolder"] = ""

# Folder files (case-insensitive)
folder_files = {f.lower() for f in os.listdir(check_folder)}

found_count = 0

for row in df.itertuples():
    found_images = []

    not_found_images = str(row.NotCopiedImage).split(",")

    for img in not_found_images:
        img_name = os.path.basename(img.strip())
        if img_name.lower() in folder_files:
            found_images.append(img_name)
            found_count += 1

    df.at[row.Index, "FoundInFolder"] = ",".join(found_images)

# Save CSV
df.to_csv(output_csv, index=False)

# Summary
print("CSV created with recovered images")
print("Total images found in folder:", found_count)
```

```python
#copying_images_from_subfolders
import os
import shutil
from collections import defaultdict

MAIN_DATASET = "/home/khushi/Pixonate/New_Anemia/R_new_anemia/Main_dataset"
OUTPUT_BASE = "/home/khushi/Pixonate/New_Anemia/R_new_anemia/work_on_data"

os.makedirs(OUTPUT_BASE, exist_ok=True)

IMAGE_EXTS = (".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff")

missing_conjunctiva = []
total_images = 0
images_per_month = defaultdict(int)

case_has_conjunctiva = {}

for root, dirs, files in os.walk(MAIN_DATASET):
    folder_name = os.path.basename(root).lower()

    # Track case folders
    if "case" in folder_name:
        case_has_conjunctiva[root] = False

    # Process conjunctiva folders anywhere
    if folder_name == "conjunctiva":

        # Mark parent case as valid
        for path in case_has_conjunctiva:
            if root.startswith(path):
                case_has_conjunctiva[path] = True

        image_files = [f for f in files if f.lower().endswith(IMAGE_EXTS)]
        count = len(image_files)
        total_images += count

        # Extract month
        month = None
        for p in root.split(os.sep):
            if "-" in p and len(p) == 7:
                month = p
                break

        for img in image_files:
            src = os.path.join(root, img)

            base, ext = os.path.splitext(img)
            dst = os.path.join(OUTPUT_BASE, img)
            i = 1
            while os.path.exists(dst):
                dst = os.path.join(OUTPUT_BASE, f"{base}_{i}{ext}")
                i += 1

            shutil.copy2(src, dst)

        if month:
            images_per_month[month] += count

# Find truly missing cases
missing_conjunctiva = [c for c, v in case_has_conjunctiva.items() if not v]

# 🔹 Output
print("\nCases WITHOUT conjunctiva folder:")
for c in missing_conjunctiva:
    print("NOT FOUND conjunctiva:", c)

print("\nSummary:")
print("Total cases missing conjunctiva:", len(missing_conjunctiva))
print("Total conjunctiva images copied:", total_images)

print("\nImages per month:")
for month, cnt in sorted(images_per_month.items()):
    print(f"{month}: {cnt}")
```
